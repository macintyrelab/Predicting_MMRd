```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
activities = pd.read_csv('./Signatures/DBsigActivities_Samples_2_filtering_YAPSAOVA+MMR+SOME.txt', sep= '\t', header=None)
```

```python
activities
```

```python
activities = activities.transpose()
activities = activities.rename(columns=activities.iloc[0])
activities = activities.iloc[1: , :]
activities
```

```python
activities['state']="MMRp"
msi_samples = ['LP16', 'LP17', 'LP18', 'LP19','LP20', 'LP21', 'LP22', 'LP24', 'LP25', 'LP26', 'LP30', 'LP37',
               'LP124', 'OVE16','OVE31', 'OVE42', 'OVE59', 'OVE64', 'OVE86', '16T131', 'MDA5', 'MDA9', 'MDA15',
               'MDA16-NG', 'MDA16']
```

```python
for msi in msi_samples:
    if msi in activities['sample'].values:
        activities.loc[activities['sample'] == msi, 'state'] = 'MMRd'
```

```python
activities
```

```python
activities.to_csv("./Signatures/Activities_DBS.txt", sep="\t", index=False)
```

```python
dbs7 = pd.DataFrame(columns = ['sample', 'DBS7', 'state'])
dbs7['sample']=activities['sample']
dbs7['DBS7']=activities['DBS7'].astype(float)
dbs7['state']=activities['state']

dbs10 = pd.DataFrame(columns = ['sample', 'DBS10', 'state'])
dbs10['sample']=activities['sample']
dbs10['DBS10']=activities['DBS10'].astype(float)
dbs10['state']=activities['state']
```

```python
import statannot

plt.rcParams["figure.figsize"] = (7,8)
fig, ax = plt.subplots()

ax = sns.boxplot(x='state', y='DBS7', data = dbs7, palette = ['royalblue', 'darkorange'], 
                 order = ['MMRp', 'MMRd'], showfliers=False, boxprops=dict(alpha=0.6))
# add stripplot
ax = sns.stripplot(x='state', y='DBS7', data = dbs7, palette = ['mediumblue', 'orangered'], 
                   jitter=0.2, size=5, order = ['MMRp', 'MMRd'])

from matplotlib.cbook import boxplot_stats
outliers_dbs7 = pd.DataFrame(columns = ['sample', 'DBS7', 'state'])
for states in dbs7['state'].unique():
        outliers = [y for stat in boxplot_stats(dbs7[dbs7['state'] == states]['DBS7']) for y in stat['fliers']]
        if outliers != []:
                for outlier in outliers:
                        outliers_dbs7 = outliers_dbs7.append(dbs7[(dbs7['state'] == states) & 
                                                                  (dbs7['DBS7'] == outlier)])
# Annotating manuallyDBS7
#ax.annotate(s=outliers_id1['sample'][27], xy=(0, outliers_id1['ID1'][27]), xytext=(-25,0), textcoords='offset points', fontsize=12)
#ax.annotate(s=outliers_id1['sample'][35], xy=(0, outliers_id1['ID1'][35]), xytext=(5,0), textcoords='offset points', fontsize=12)
#ax.annotate(s=outliers_id1['sample'][38], xy=(0, outliers_id1['ID1'][38]), xytext=(5,-10), textcoords='offset points', fontsize=12)
#ax.annotate(s=outliers_id1['sample'][86], xy=(0, outliers_id1['ID1'][86]), xytext=(5,0), textcoords='offset points', fontsize=12)
#ax.annotate(s=outliers_id1['sample'][89], xy=(0, outliers_id1['ID1'][89]), xytext=(5,0), textcoords='offset points', fontsize=12)
#ax.annotate(s=outliers_id1['sample'][95], xy=(0, outliers_id1['ID1'][95]), xytext=(5,0), textcoords='offset points', fontsize=12)
#ax.annotate(s=outliers_id1['sample'][125], xy=(0, outliers_id1['ID1'][125]), xytext=(-30,0), textcoords='offset points', fontsize=12)

ax.set_xlabel("Clinical MMR status",fontsize=13)
ax.set_ylabel("Activity of DBS7",fontsize=13)
ax.tick_params(labelsize=12)

import statannot
statannot.add_stat_annotation(
    ax,
    data=dbs7,
    x='state',
    y='DBS7',
    box_pairs=[("MMRp", "MMRd")],
    order = ['MMRp', 'MMRd'],
    test="Mann-Whitney",
    text_format="star",
    loc="inside")


# show the graph
plt.title("DBS7", fontsize = 14)
plt.savefig("./Signatures/dbs7.png")
plt.show()
plt.draw()
```
