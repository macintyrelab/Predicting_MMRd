```python
old_data_1 = pd.read_csv('./TMB_1/OVAseq_old_tmb_VC_3_10%.txt', sep= '\t')
new_data_1 = pd.read_csv('./TMB_1/OVAseq_new_tmb_VC_3_10%.txt', sep= '\t')
```

```python
tmb_old_1 = old_data_1.pivot_table(columns=['sampleID'], aggfunc='size')
tmb_new_1 = new_data_1.pivot_table(columns=['sampleID'], aggfunc='size')
```

```python
tmb_old_1 = tmb_old_1*1000000/863493
tmb_new_1 = tmb_new_1*1000000/921145
data_tmb = tmb_old_1.append(tmb_new_1)
```

```python
tmb_both = pd.DataFrame(columns = ['tmb', 'State'])
tmb_both['tmb'] = data_tmb
```

```python
tmb_both['State']="MMRp"
```

```python
msi_samples = ['LP16', 'LP17', 'LP18', 'LP19','LP20', 'LP21', 'LP22', 'LP24', 'LP25', 'LP26', 'LP30', 'LP37',
               'LP124', 'OVE16','OVE31', 'OVE42', 'OVE59', 'OVE64', 'OVE86', '16T131', 'MDA16', 'MDA15', 'MDA9','MDA16']
```

```python
tmb_both = tmb_both.reset_index()
tmb_both
```

```python
for msi in msi_samples:
    if msi in tmb_both['sampleID'].values:
        tmb_both.loc[tmb_both['sampleID'] == msi, 'State'] = 'MMRd'
```

```python
import numpy as np
import statannot
plt.rcParams["figure.figsize"] = (6.5,9)

ax = sns.boxplot(x='State', y='tmb', data=tmb_both, palette = ['royalblue', 'darkorange'], 
                 showfliers=False, boxprops=dict(alpha=0.6), order = ['MMRp', 'MMRd'])
# add stripplot
np.random.seed(10)
ax = sns.stripplot(x='State', y='tmb', data=tmb_both, color="orange", 
                   jitter=0.2, size=5, palette = ['mediumblue', 'orangered'], order = ['MMRp', 'MMRd'])
# Adding statistical significance

# Labelling outliers
from matplotlib.cbook import boxplot_stats
outliers_df = pd.DataFrame(columns = ['sampleID', 'tmb', 'State'])
for state in tmb_both['State'].unique():
        outliers = [y for stat in boxplot_stats(tmb_both[tmb_both['State'] == state]['tmb']) for y in stat['fliers']]
        if outliers != []:
                for outlier in outliers:
                        outliers_df = outliers_df.append(tmb_both[(tmb_both['State'] == state) & (tmb_both['tmb'] == outlier)])

# Annotating manually
ax.annotate(s=outliers_df['sampleID'][28], xy=(0, outliers_df['tmb'][28]), xytext=(-40,0), textcoords='offset points', fontsize=14)
ax.annotate(s=outliers_df['sampleID'][45], xy=(0, outliers_df['tmb'][45]), xytext=(-13,-13), textcoords='offset points', fontsize=14)
ax.annotate(s=outliers_df['sampleID'][70], xy=(0, outliers_df['tmb'][70]), xytext=(37,7), textcoords='offset points', fontsize=14)
ax.annotate(s=outliers_df['sampleID'][76], xy=(0, outliers_df['tmb'][76]), xytext=(-33,0), textcoords='offset points', fontsize=14)
ax.annotate(s=outliers_df['sampleID'][83], xy=(0, outliers_df['tmb'][83]), xytext=(-30,3), textcoords='offset points', fontsize=14)
ax.annotate(s=outliers_df['sampleID'][114], xy=(0, outliers_df['tmb'][114]), xytext=(8,3), textcoords='offset points', fontsize=14)
ax.annotate(s=outliers_df['sampleID'][123], xy=(0, outliers_df['tmb'][123]), xytext=(8,-2), textcoords='offset points', fontsize=14)
ax.annotate(s=outliers_df['sampleID'][126], xy=(0, outliers_df['tmb'][126]), xytext=(-60,15), textcoords='offset points', fontsize=14)
ax.annotate(s=tmb_both['sampleID'][13], xy=(1, tmb_both['tmb'][13]), xytext=(15,-7), textcoords='offset points', fontsize=14, color = 'dimgray')
ax.annotate(s=tmb_both['sampleID'][14], xy=(1, tmb_both['tmb'][14]), xytext=(20,-15), textcoords='offset points', fontsize=14, color = 'dimgray')
ax.annotate(s=tmb_both['sampleID'][59], xy=(1, tmb_both['tmb'][59]), xytext=(-35,-7), textcoords='offset points', fontsize=14, color = 'dimgray')
ax.annotate(s=tmb_both['sampleID'][102], xy=(1, tmb_both['tmb'][102]), xytext=(-90,-12), textcoords='offset points', fontsize=14, color = 'dimgray')
ax.annotate(s=tmb_both['sampleID'][103], xy=(1, tmb_both['tmb'][103]), xytext=(-13,-13), textcoords='offset points', fontsize=14)

ax.set_xlabel("Clinical MMR status",fontsize=15)
ax.set_ylabel("Nonsynonymous mutations per Mb",fontsize=15)
ax.tick_params(labelsize=14)


statannot.add_stat_annotation(
    ax,
    data=tmb_both,
    x='State',
    y='tmb',
    box_pairs=[("MMRp", "MMRd")],
    order = ['MMRp', 'MMRd'],
    test="Mann-Whitney",
    text_format="star",
    loc="inside")
plt.title("TMB", fontsize =16)
plt.savefig('./TMB_10%_stripplot_poster.png')
```
