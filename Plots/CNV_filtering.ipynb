```python
import os
import pandas as pd
import numpy as np
cnv_data = pd.DataFrame(columns=['sample','cnv'])
```

```python
directory = '/home/osboxes/MasterGitHub/DGL/python/CNV/Profiles_old/'
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    sample = filename.split('_')[0]
    if os.path.isfile(f):
        df = pd.read_csv(f, sep="\t")
        df_final = df.drop(df.index[(df['total_copy_number']==2)])
        cnv_data.at[filename,"sample"]=sample
        cnv = len(df_final.index)
        cnv_data.at[filename,"cnv"]=float(cnv)
```

```python
import pandas as pd
copy_number = pd.read_csv("./CNV/Results/CNV.tsv", sep="\t")
copy_number.head()
```

```python
copy_number['State']='MMRp'
```

```python
msi_samples = ['LP16', 'LP17', 'LP18', 'LP19','LP20', 'LP21', 'LP22', 'LP24', 'LP25', 'LP26', 'LP30', 'LP37',
               'LP124', 'OVE16','OVE31', 'OVE42', 'OVE59', 'OVE64', 'OVE86', '16T131', 'MDA5', 'MDA9', 'MDA15',
               'MDA16-NG', 'MDA16']
```

```python
for msi in msi_samples:
    if msi in copy_number['sample'].values:
        copy_number.loc[copy_number['sample'] == msi, 'State'] = 'MMRd'
```

```python
copy_number#.loc[copy_number['state']=="MMRd"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
order_list = ['MMRp', 'MMRd']
sns.violinplot(x='State', y='cnv', data=copy_number, color="0.8", cut=0, order=order_list)
sns.stripplot(x='State', y='cnv', data=copy_number, jitter=True, order=order_list,
              palette = ['steelblue','darkorange'])
plt.show()
plt.savefig('./CNV/CNV_plot.png')
```

```python
import statannot
import numpy as np

copy_number = copy_number.dropna()
order_list = ['MMRp', 'MMRd']
plt.rcParams["figure.figsize"] = (6.5,9)
fig, ax = plt.subplots()
ax = sns.boxplot(x='State', y='cnv', data=copy_number, order=order_list, 
                 palette = ['royalblue', 'darkorange'], showfliers=False, boxprops=dict(alpha=0.6))
# add stripplot
np.random.seed(1)
ax = sns.stripplot(x='State', y='cnv', data=copy_number, order=order_list, 
                   jitter=0.2, size=5, palette = ['mediumblue', 'orangered'])

# Labelling outliers
from matplotlib.cbook import boxplot_stats
outliers_df = pd.DataFrame(columns = ['sample', 'cnv', 'State'])
for state in copy_number['State'].unique():
        outliers = [y for stat in boxplot_stats(copy_number[copy_number['State'] == state]['cnv']) for y in stat['fliers']]
        if outliers != []:
                for outlier in outliers:
                        outliers_df = outliers_df.append(copy_number[(copy_number['State'] == state) & (copy_number['cnv'] == outlier)])

# Annotating manually
ax.annotate(s=outliers_df['sample'][14], xy=(1, outliers_df['cnv'][14]), xytext=(-7,0), textcoords='offset points', fontsize=14)
ax.annotate(s=outliers_df['sample'][15], xy=(1, outliers_df['cnv'][15]), xytext=(-25,0), textcoords='offset points', fontsize=14)
#ax.annotate(s=outliers_df['sample'][66], xy=(0, outliers_df['cnv'][66]), xytext=(-57,-13), textcoords='offset points', fontsize=14)
#ax.annotate(s=outliers_df['sample'][92], xy=(0, outliers_df['cnv'][92]), xytext=(-3,3), textcoords='offset points', fontsize=14)

ax.set_xlabel("Clinical MMR status",fontsize=15)
ax.set_ylabel("Number of copy unmber events",fontsize=15)
ax.tick_params(labelsize=13.5)

# Annotating outliers
#i=3
#j=30
#for row in outliers_df.iterrows():
#        i=i*(-1)
#        j=j*(-1)
#        if row[1]['State'] == "MMRp":
#            ax.annotate(s=row[1]['sample'], xy=(0, row[1]['%']), xytext=(j,i), textcoords='offset points', fontsize=12)
#        else:
#            ax.annotate(s=row[1]['sample'], xy=(2, row[1]['%']), xytext=(j,i), textcoords='offset points', fontsize=12)

# Adding statistical significance
statannot.add_stat_annotation(
    ax,
    data=copy_number,
    x='State',
    y='cnv',
    box_pairs=[("MMRp", "MMRd")],
    order = ['MMRp', 'MMRd'],
    test="Mann-Whitney",
    text_format="star",
    loc="inside")

# Adding plot title
plt.title("CNV", fontsize = 16)
plt.savefig('./CNV/CNV_boxplot_poster.png', dpi = 1200)
```
