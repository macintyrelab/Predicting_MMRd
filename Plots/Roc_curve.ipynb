```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8,7.5)
y_gs = [0, 0.8, 1]
x_gs = [0, 0.041, 1]

y_lg = [0, 1, 1]
x_lg = [0, 0.0164, 1]

plt.plot([0, 1], [0, 1], "k--")
plt.plot(x_gs, y_gs, "-",  label = "Gold standard techniques")
plt.plot(x_lg, y_lg, "-",  label = "Linear regression classifier")

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel("True Positive rate", fontsize = 16)
plt.ylabel("False Positive Rate", fontsize = 16)
plt.tick_params(labelsize=14)
plt.legend(loc='lower right', fontsize = 15)
plt.savefig("RocCurve.png", dpi=600);
```
