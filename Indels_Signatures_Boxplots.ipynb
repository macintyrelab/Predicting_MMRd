```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
activities = pd.read_csv('./Signatures/IDsigActivities_Samples_2_filtering_YAPSAOVA+MMR+SOME.txt', sep= '\t', header=None)
```

```python
activities
```

```python
activities = activities.transpose()
activities = activities.rename(columns=activities.iloc[0])
activities = activities.iloc[1: , :]
activities
```

```python
activities['state']="MMRp"
msi_samples = ['LP16', 'LP17', 'LP18', 'LP19','LP20', 'LP21', 'LP22', 'LP24', 'LP25', 'LP26', 'LP30', 'LP37',
               'LP124', 'OVE16','OVE31', 'OVE42', 'OVE59', 'OVE64', 'OVE86', '16T131', 'MDA5', 'MDA9', 'MDA15',
               'MDA16-NG', 'MDA16']
```

```python
for msi in msi_samples:
    if msi in activities['sample'].values:
        activities.loc[activities['sample'] == msi, 'state'] = 'MMRd'
```

```python
activities
```

```python
activities.to_csv("./Signatures/Activities_indels.txt", sep="\t", index=False)
```

```python
activities = pd.read_csv('./Signatures/Activities_indels.txt', sep= '\t')
activities
```

```python
id1 = pd.DataFrame(columns = ['sample', 'ID1', 'state'])
id1['sample']=activities['sample']
id1['ID1']=activities['ID1']
id1['state']=activities['state']

id2 = pd.DataFrame(columns = ['sample', 'ID2', 'state'])
id2['sample']=activities['sample']
id2['ID2']=activities['ID2']
id2['state']=activities['state']

id7 = pd.DataFrame(columns = ['sample', 'ID7', 'state'])
id7['sample']=activities['sample']
id7['ID7']=activities['ID7']
id7['state']=activities['state']

id2_7 = pd.DataFrame(columns = ['sample', 'ID2+ID7', 'state'])
id2_7['sample']=activities['sample']
id2_7['ID2+ID7']=activities['ID2'] + activities['ID7']
id2_7['state']=activities['state']
id2_7.to_csv("ID2+ID7.txt", sep ="\t", index=False)
```

```python
id2_7
```

```python
import statannot
plt.rcParams["figure.figsize"] = (7,8)
fig, ax = plt.subplots()

ax = sns.boxplot(x='state', y='ID1', data = id1, palette = ['royalblue', 'darkorange'], 
                 order = ['MMRp', 'MMRd'], showfliers = False, boxprops=dict(alpha=0.6))
# add stripplot
ax = sns.stripplot(x='state', y='ID1', data = id1, palette = ['mediumblue', 'orangered'], 
                   jitter=0.2, size=5, order = ['MMRp', 'MMRd'])

from matplotlib.cbook import boxplot_stats
outliers_id1 = pd.DataFrame(columns = ['sample', 'ID1', 'state'])
for states in id1['state'].unique():
        outliers = [y for stat in boxplot_stats(id1[id1['state'] == states]['ID1']) for y in stat['fliers']]
        if outliers != []:
                for outlier in outliers:
                        outliers_id1 = outliers_id1.append(id1[(id1['state'] == states) & (id1['ID1'] == outlier)])

ax.set_xlabel("Clinical MMR status",fontsize=13)
ax.set_ylabel("Activity of ID1",fontsize=13)
ax.tick_params(labelsize=12)

import statannot
statannot.add_stat_annotation(
    ax,
    data=id1,
    x='state',
    y='ID1',
    box_pairs=[("MMRp", "MMRd")],
    order = ['MMRp', 'MMRd'],
    test="Mann-Whitney",
    text_format="star",
    loc="inside")

# add title
plt.title("ID1", fontsize = 14)
# show the graph
plt.savefig("./Signatures/ID1.png")
plt.show()
plt.draw()
```

```python
id2_mix = pd.DataFrame(columns = ['sample', 'value', 'indel', 'state'])
id2_mix['sample']=activities['sample']
id2_mix['value']=activities['ID2']
id2_mix['indel']="ID2"
id2_mix['state']=activities['state']

id7_mix = pd.DataFrame(columns = ['sample', 'value', 'indel', 'state'])
id7_mix['sample']=activities['sample']
id7_mix['value']=activities['ID7']
id7_mix['indel']="ID7"
id7_mix['state']=activities['state']

id_mix = id2_mix.append(id7_mix)
```

```python
id_mix = id_mix.sort_values('indel', ascending= True)
```

```python
plt.rcParams["figure.figsize"] = (6.5,7)
# make grouped boxplot and save it in a variable
bp = sns.boxplot(y='value', x='indel', 
                 data=id_mix, 
                 palette = ['royalblue', 'darkorange'],
                 hue = 'state', 
                 showfliers = False,
                 boxprops=dict(alpha=0.6))
 
# make grouped stripplot and save it in a variable
bp = sns.stripplot(y='value', x='indel', 
                   data=id_mix, 
                   jitter=True,
                   dodge=True, 
                   hue = 'state',
                   palette = ['mediumblue', 'orangered'])
# get legend information from the plot object
handles, labels = bp.get_legend_handles_labels()

import statannot
statannot.add_stat_annotation(
    ax = sns.boxplot(data=id_mix, x='indel', y='value', hue='state', boxprops=dict(alpha=0),
                     palette = ['royalblue', 'darkorange'], showfliers = False),
    hue = 'state',
    data=id_mix,
    x='indel',
    y='value',
    box_pairs=[(('ID2',"MMRp"), ('ID2',"MMRd")), (('ID7',"MMRp"), ('ID7',"MMRd"))],
    test="Mann-Whitney",
    text_format="star",
    loc="inside")

plt.ylabel('Signature activity', fontsize = 15)
plt.xlabel('Clinical MMR status', fontsize = 15)
#plt.title('Indel mutational signatures', fontsize = 15)
plt.tick_params(labelsize=13)

l = plt.legend(handles[0:2], labels[0:2], bbox_to_anchor=(0.7,-0.15), fontsize = 13)
plt.savefig("./Signatures/ID2+ID7_boxplots.png")
```
