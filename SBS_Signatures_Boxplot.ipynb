```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
activities = pd.read_csv('./Signatures/SBsigActivities_Samples_2_filtering_YAPSAOVA+MMR+SOME.txt', sep= '\t', header=None)
```

```python
activities
```

```python
activities = activities.transpose()
activities = activities.rename(columns=activities.iloc[0])
activities = activities.iloc[1: , :]
activities
```

```python
activities['state']="MMRp"
msi_samples = ['LP16', 'LP17', 'LP18', 'LP19','LP20', 'LP21', 'LP22', 'LP24', 'LP25', 'LP26', 'LP30', 'LP37',
               'LP124', 'OVE16','OVE31', 'OVE42', 'OVE59', 'OVE64', 'OVE86', '16T131', 'MDA5', 'MDA9', 'MDA15',
               'MDA16-NG', 'MDA16']
```

```python
for msi in msi_samples:
    if msi in activities['sample'].values:
        activities.loc[activities['sample'] == msi, 'state'] = 'MMRd'
```

```python
activities
```

```python
activities.to_csv("./Signatures/Activities_SBS.txt", sep="\t", index=False)
```

```python
SBS6 = pd.DataFrame(columns = ['sample', 'SBS6', 'state'])
SBS6['sample']=activities['sample']
SBS6['SBS6']=activities['SBS6'].astype(float)
SBS6['state']=activities['state']

SBS14 = pd.DataFrame(columns = ['sample', 'SBS14', 'state'])
SBS14['sample']=activities['sample']
SBS14['SBS14']=activities['SBS14'].astype(float)
SBS14['state']=activities['state']

SBS15 = pd.DataFrame(columns = ['sample', 'SBS15', 'state'])
SBS15['sample']=activities['sample']
SBS15['SBS15']=activities['SBS15'].astype(float)
SBS15['state']=activities['state']

SBS20 = pd.DataFrame(columns = ['sample', 'SBS20', 'state'])
SBS20['sample']=activities['sample']
SBS20['SBS20']=activities['SBS20'].astype(float) 
SBS20['state']=activities['state']

SBS21 = pd.DataFrame(columns = ['sample', 'SBS21', 'state'])
SBS21['sample']=activities['sample']
SBS21['SBS21']=activities['SBS21'].astype(float)
SBS21['state']=activities['state']

SBS26 = pd.DataFrame(columns = ['sample', 'SBS26', 'state'])
SBS26['sample']=activities['sample']
SBS26['SBS26']=activities['SBS26'].astype(float)
SBS26['state']=activities['state']

SBS44 = pd.DataFrame(columns = ['sample', 'SBS44', 'state'])
SBS44['sample']=activities['sample']
SBS44['SBS44']=activities['SBS44'].astype(float) 
SBS44['state']=activities['state']
```

```python
import statannot
plt.rcParams["figure.figsize"] = (7,8)
fig, ax = plt.subplots()

ax = sns.boxplot(x='state', y='SBS6', data = SBS6, palette = ['royalblue', 'darkorange'], 
                 order = ['MMRp', 'MMRd'], showfliers=False, boxprops=dict(alpha=0.6))
# add stripplot
ax = sns.stripplot(x='state', y='SBS6', data = SBS6, palette = ['mediumblue', 'orangered'], 
                   jitter=0.2, size=5, order = ['MMRp', 'MMRd'])

from matplotlib.cbook import boxplot_stats
outliers_SBS6 = pd.DataFrame(columns = ['sample', 'SBS6', 'state'])
for states in SBS6['state'].unique():
        outliers = [y for stat in boxplot_stats(SBS6[SBS6['state'] == states]['SBS6']) for y in stat['fliers']]
        if outliers != []:
                for outlier in outliers:
                        outliers_SBS6 = outliers_SBS6.append(SBS6[(SBS6['state'] == states) & 
                                                                  (SBS6['SBS6'] == outlier)])
# Annotating manually
#ax.annotate(s=outliers_id1['sample'][27], xy=(0, outliers_id1['ID1'][27]), xytext=(-25,0), textcoords='offset points', fontsize=12)
#ax.annotate(s=outliers_id1['sample'][35], xy=(0, outliers_id1['ID1'][35]), xytext=(5,0), textcoords='offset points', fontsize=12)
#ax.annotate(s=outliers_id1['sample'][38], xy=(0, outliers_id1['ID1'][38]), xytext=(5,-10), textcoords='offset points', fontsize=12)
#ax.annotate(s=outliers_id1['sample'][86], xy=(0, outliers_id1['ID1'][86]), xytext=(5,0), textcoords='offset points', fontsize=12)
#ax.annotate(s=outliers_id1['sample'][89], xy=(0, outliers_id1['ID1'][89]), xytext=(5,0), textcoords='offset points', fontsize=12)
#ax.annotate(s=outliers_id1['sample'][95], xy=(0, outliers_id1['ID1'][95]), xytext=(5,0), textcoords='offset points', fontsize=12)
#ax.annotate(s=outliers_id1['sample'][125], xy=(0, outliers_id1['ID1'][125]), xytext=(-30,0), textcoords='offset points', fontsize=12)

ax.set_xlabel("Clinical MMR status",fontsize=13)
ax.set_ylabel("Activity of SBS6",fontsize=13)
ax.tick_params(labelsize=12)

import statannot
statannot.add_stat_annotation(
    ax,
    data=SBS6,
    x='state',
    y='SBS6',
    box_pairs=[("MMRp", "MMRd")],
    order = ['MMRp', 'MMRd'],
    test="Mann-Whitney",
    text_format="star",
    loc="inside")

# add title
plt.title("SBS6", fontsize = 14)

# show the graph
plt.savefig("./Signatures/SBS6.png")
plt.show()
plt.draw()
```
