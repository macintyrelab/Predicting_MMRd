```python
import pandas as pd
df = pd.read_csv("PCR_MSI_Results.txt", sep="\t")
df.head()
```

```python
database = pd.DataFrame(columns=['Sample', 'HISTOTYPE','PCR', 'IHQ', 'DIAGNOSIS', 'NGS_MMRd', 'MSI', 'TMB', 
                                 'ID_Signature', 'CNV'])
database['Sample'] = df['ID_CNIO ']
database['PCR'] = df['CLASIFICACION']
database['IHQ'] = df['IHQ_PROTEINA_CANDIDATA ']
database['DIAGNOSIS']="MMRp"
database['NGS_MMRd']=0
database
```

```python
# Adding tumor type
import pandas as pd
Subtype=open("DB_Cancer_type.txt")
Subtype_lines=Subtype.readlines()
SubtypeDic={}
for i in Subtype_lines:
    Subtype_fields=i.split('\t')
    SubtypeDic[Subtype_fields[0]]=Subtype_fields[1].rstrip('\n')
Subtype.close()

for sample in database["Sample"]:
    if sample in SubtypeDic:
        database.loc[database.Sample == sample, 'HISTOTYPE'] = SubtypeDic[sample]
    else:
        print(sample)
        database = database.drop(database.index[(database['Sample'] == sample)]) 
```

```python
# Checking the samples from the database that are present in the MSI_total.txt file
import pandas as pd
MSI=open("MSI_total.txt")
MSI_lines=MSI.readlines()
MSIDic={}
for i in MSI_lines:
    MSI_fields=i.split('\t')
    MSIDic[MSI_fields[0]]=MSI_fields[1].rstrip('\n')
MSI.close()

for sample in database["Sample"]:
    if sample in MSIDic:
        database.loc[database.Sample == sample, 'MSI'] = MSIDic[sample]
    else:
        print(sample)
        database = database.drop(database.index[(database['Sample'] == sample)]) 
```

```python
# Checking that all samples from MSI_total.txt are in the database
import pandas as pd
PCR_MSI=open("PCR_MSI_Results.txt")
PCR_MSI_lines=PCR_MSI.readlines()
DataDic={}
for i in PCR_MSI_lines:
    PCR_MSI_fields=i.split('\t')
    DataDic[PCR_MSI_fields[0]]=PCR_MSI_fields[1].rstrip('\n')
PCR_MSI.close()

data_frame = pd.read_csv("MSI_total.txt", sep="\t")
for samp in data_frame["sample"]:
    if not samp in DataDic:
        print(samp)
```
